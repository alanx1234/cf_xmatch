```python
import pandas as pd
import sys
sys.path.insert(0, '.')
from chronoflow.ChronoFlow import ChronoFlow
import numpy as np

fgkm = pd.read_csv('cf_data/pass2022_fgkm.csv')

# need logCerr — repull BP/RP fluxes for companions with prot_days
from astroquery.gaia import Gaia

companions = fgkm[(fgkm['role'] == 'companion') & fgkm['prot_days'].notna()].copy()
ids_str = ', '.join(str(i) for i in companions['source_id'].astype('Int64').tolist())

flux_results = Gaia.launch_job(f"""
    SELECT source_id, phot_bp_mean_flux, phot_bp_mean_flux_error,
           phot_rp_mean_flux, phot_rp_mean_flux_error
    FROM gaiadr3.gaia_source
    WHERE source_id IN ({ids_str})
""").get_results().to_pandas()

flux_results['source_id'] = flux_results['source_id'].astype('Int64')
flux_results['sigma_bprp'] = (2.5 / np.log(10)) * np.sqrt(
    (flux_results['phot_bp_mean_flux_error'] / flux_results['phot_bp_mean_flux'])**2 +
    (flux_results['phot_rp_mean_flux_error'] / flux_results['phot_rp_mean_flux'])**2
)
flux_results['logCerr'] = np.log10(flux_results['sigma_bprp'])

companions = companions.merge(
    flux_results[['source_id', 'logCerr']],
    on='source_id',
    how='left'
)
companions['logCerr'] = companions['logCerr'].fillna(-1.55)

fgk = companions.copy()
print(f'running ChronoFlow on {len(fgk)} FGK companions')
print(fgk[['name', 'prot_days', 'bprp0', 'logCerr']])

# ChronoFlow inference
WEIGHTS_BPRP = 'model_output_BPRP_0/weights.pth'
cf_model = ChronoFlow(weights_file=WEIGHTS_BPRP)
logA_grid = np.linspace(0, 4.14, 1000)

results = []
for _, row in fgk.iterrows():
    post = cf_model.calcPostAge(
        logProt=np.log10(float(row['prot_days'])),
        BPRP0=float(row['bprp0']),
        logCerr=float(row['logCerr'])
    )
    medLogA, logA_Err, logA_err = cf_model.getAgeSummStats(post, logA_grid)
    age_gyr = 10**medLogA / 1000
    age_gyr_upper = (10**(medLogA + logA_Err) / 1000) - age_gyr
    age_gyr_lower = age_gyr - (10**(medLogA + logA_err) / 1000)
    results.append({
        'name': row['name'],
        'prot_days': row['prot_days'],
        'bprp0': row['bprp0'],
        'medLogA_Myr': medLogA,
        'age_gyr_cf': age_gyr,
        'age_gyr_upper': age_gyr_upper,
        'age_gyr_lower': age_gyr_lower
    })

results_df = pd.DataFrame(results)
print(results_df)
```

```python
fgkm = pd.read_csv('cf_data/pass2022_fgkm.csv')

# update companion rows with new cf ages
fgkm = fgkm.merge(
    results_df[['name', 'age_gyr_cf', 'age_gyr_upper', 'age_gyr_lower']],
    on='name',
    how='left',
    suffixes=('_old', '')
)
fgkm = fgkm.drop(columns=['age_gyr_cf_old', 'age_gyr_upper_old', 'age_gyr_lower_old'])

# propagate companion ages to their paired primary (always the row above)
for i in range(1, len(fgkm), 2):  # companion rows are odd indices
    for col in ['age_gyr_cf', 'age_gyr_upper', 'age_gyr_lower']:
        val = fgkm.at[i, col]
        if pd.notna(val):
            fgkm.at[i-1, col] = val

fgkm.to_csv('cf_data/pass2022_fgkm.csv', index=False)
print('saved!')
fgkm[['name', 'role', 'age_gyr_cf', 'age_gyr_upper', 'age_gyr_lower']]
```
