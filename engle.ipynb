```python
engle = pd.read_csv("cf_data/engle_unified_long_from_pdfs.csv")

cols = [c for c in engle.columns if any(k in c.lower() for k in ["prot", "age", "err", "unc"])]
print(cols)

display(engle[cols].head(10))
```

```python
engle_clean = engle.copy()

# age
if "age_gyr_err_plus" in engle_clean.columns:
    engle_clean["age_err_hi_gyr"] = pd.to_numeric(engle_clean["age_gyr_err_plus"], errors="coerce").abs()

if "age_gyr_err_minus" in engle_clean.columns:
    engle_clean["age_err_lo_gyr"] = pd.to_numeric(engle_clean["age_gyr_err_minus"], errors="coerce").abs()

# prot
if "prot_days_err_plus" in engle_clean.columns:
    engle_clean["prot_err_hi_days"] = pd.to_numeric(engle_clean["prot_days_err_plus"], errors="coerce").abs()

if "prot_days_err_minus" in engle_clean.columns:
    engle_clean["prot_err_lo_days"] = pd.to_numeric(engle_clean["prot_days_err_minus"], errors="coerce").abs()

# log age, if present
if "log_age_yr_err_plus" in engle_clean.columns:
    engle_clean["log_age_err_hi"] = pd.to_numeric(engle_clean["log_age_yr_err_plus"], errors="coerce").abs()

if "log_age_yr_err_minus" in engle_clean.columns:
    engle_clean["log_age_err_lo"] = pd.to_numeric(engle_clean["log_age_yr_err_minus"], errors="coerce").abs()
```

```python
cols = [c for c in [
    "star_name",
    "source_paper",
    "source_table",
    "age_gyr_raw",
    "age_gyr",
    "age_gyr_err_plus",
    "age_gyr_err_minus",
    "age_err_lo_gyr",
    "age_err_hi_gyr",
    "prot_days_raw",
    "prot_days",
    "prot_days_err_plus",
    "prot_days_err_minus",
    "prot_err_lo_days",
    "prot_err_hi_days",
] if c in engle_clean.columns]

display(engle_clean[cols].head(15))
```

```python
import re
import pandas as pd
import numpy as np

engle_clean = engle_clean.copy()

def extract_trailing_note(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    m = re.search(r'([a-zA-Z]+)\s*$', s)
    return m.group(1) if m else np.nan

engle_clean["prot_note_ref"] = engle_clean["prot_days_raw"].apply(extract_trailing_note) if "prot_days_raw" in engle_clean.columns else np.nan
engle_clean["age_note_ref"] = engle_clean["age_gyr_raw"].apply(extract_trailing_note) if "age_gyr_raw" in engle_clean.columns else np.nan

display(
    engle_clean[
        [c for c in ["star_name", "prot_days_raw", "prot_note_ref", "age_gyr_raw", "age_note_ref"] if c in engle_clean.columns]
    ].head(15)
)
```
